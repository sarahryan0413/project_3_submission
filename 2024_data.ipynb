```python
#Import Dependencies

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import CSV file
file_path = 'data/reports_by_month.csv'
sighting_df = pd.read_csv(file_path)
monthly_2024_df=sighting_df.iloc[4:16, 0:3]
monthly_2024_df.head()
```

```python
monthly_2024_df.plot.bar(x= 'sighting_month', y='count')
plt.title("Number of Sightings by Month")
plt.tight_layout()

plt.show()
```

```python
csv2= 'data/sightings_2024.csv'
country_df = pd.read_csv(csv2)
country_counts = country_df["country"].value_counts().sort_values(ascending=False)
print(country_counts)
```

```python
country_counts_filtered = country_counts[country_counts > 20]
print(country_counts_filtered)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(country_counts_filtered, labels=country_counts_filtered.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("UFO Sightings by Country 2024", fontsize=8)
plt.show()
```
